```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Load the CSV file with user URLs
df = pd.read_csv("User_Url.csv")
print(df)

# Initialize an empty list to store rating URLs
url_of_ratings = []

# Define headers to mimic a browser (optional, but can help with requests)
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
}

```

```python
# Loop through the list of URLs
for url in df['0']:
        # Request the page content
    response = requests.get(url, headers=headers)
        
        # Parse the page content with BeautifulSoup
    soup = BeautifulSoup(response.content, 'html.parser')
        
        # Find the div with class 'profilePageUserStatsInfo' and extract the link to ratings
    profile_info = soup.find('div', class_='profilePageUserStatsInfo')
    print(profile_info)
        
```

```python
if profile_info:
            rating_url = profile_info.find('a', href=True)['href']  # Extract the href link
            full_rating_url = "https://www.goodreads.com" + rating_url  # Append base URL to the href
            
            # Append the rating URL to the list
            url_of_ratings.append(full_rating_url)
        else:
            print(f"Ratings not found for {url}")
    except Exception as e:
        print(f"Error processing {url}: {e}")

# Output the list of rating URLs
print(url_of_ratings)
```
