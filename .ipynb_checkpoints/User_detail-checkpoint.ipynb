```python
import nest_asyncio
import asyncio
from playwright.async_api import async_playwright

# Apply the nested asyncio patch
nest_asyncio.apply()

async def main():
    async with async_playwright() as p:
        # Launch a browser
        browser = await p.chromium.launch(headless=False)  # Use headless=True to run without GUI
        context = await browser.new_context()

        # Open a new page and navigate to the user profile page
        page = await context.new_page()
        await page.goto("https://www.goodreads.com/user/show/107704225")

        # Wait for necessary elements to load
        await page.wait_for_selector(".mainContent")

        # Extract the information you need
        profile_content = await page.locator(".mainContent").inner_html()
        print(profile_content)

        # Close the browser
        await browser.close()

# Run the async function
await main()
```

```python
import nest_asyncio
import asyncio
from playwright.async_api import async_playwright

# Apply the nested asyncio patch
nest_asyncio.apply()

async def main():
    async with async_playwright() as p:
        # Launch a browser
        browser = await p.chromium.launch(headless=False)  # Use headless=True to run without GUI
        context = await browser.new_context()

        # Open a new page
        page = await context.new_page()

        # Navigate to the Goodreads login page
        await page.goto("https://www.goodreads.com/ap/signin?language=en_US&openid.assoc_handle=amzn_goodreads_web_na&openid.claimed_id=http%3A%2F%2Fspecs.openid.net%2Fauth%2F2.0%2Fidentifier_select&openid.identity=http%3A%2F%2Fspecs.openid.net%2Fauth%2F2.0%2Fidentifier_select&openid.mode=checkid_setup&openid.ns=http%3A%2F%2Fspecs.openid.net%2Fauth%2F2.0&openid.pape.max_auth_age=0&openid.return_to=https%3A%2F%2Fwww.goodreads.com%2Fap-handler%2Fsign-in&siteState=eyJyZXR1cm5fdXJsIjoiaHR0cHM6Ly93d3cuZ29vZHJlYWRzLmNvbS8ifQ%3D%3D")

        # Wait for the email input to appear
        await page.wait_for_selector("input[name='user[email]']", timeout=5000)  # Wait for email field for up to 5 seconds

        # Fill in the login form
        #await page.fill("input[name='user[email]']", "robiulriyadh66@gmail.com")  # Replace with your email
        await page.fill("input#ap_email", "robiulriyadh66@gmail.com")
        await page.fill("input[name='user[password]']", "V!MZUhP8TL39-jH")        # Replace with your password
        await page.click("input[type='submit']")

        # Wait for login completion (e.g., wait for an element that is visible after logging in)
        await page.wait_for_selector(".siteHeader__topLevelLink", timeout=10000)  # Wait for a logged-in element

        # Navigate to the restricted user profile
        await page.goto("https://www.goodreads.com/user/show/107704225")

        # Wait for necessary elements to load
        await page.wait_for_selector(".mainContent", timeout=10000)

        # Extract the information you need
        profile_content = await page.locator(".mainContent").inner_html()
        print(profile_content)

        # Close the browser
        await browser.close()

# Run the async function
await main()
```
